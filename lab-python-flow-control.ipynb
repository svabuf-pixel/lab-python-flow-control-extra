*Introduction to Functions*:

    Functions are blocks of code that perform a specific task. They are used to break up complex code into smaller, more manageable parts, which can make your code easier to read and understand. Functions can also be reused multiple times throughout your code, which can save you a lot of time and effort.

    Functions are defined using the def keyword, followed by the name of the function and a set of parentheses. Inside the parentheses, you can list any arguments that the function needs in order to perform its task. These arguments are optional, but they can be useful if you need to pass data into the function from outside.

    Once you have defined a function, you can call it from anywhere in your code using its name and passing any necessary arguments. When the function is called, the code inside it is executed, and any values that it returns are passed back to the calling code.

```python
import random

# Inventory (Cross, Shield, Holy Water, Salt)
```

```python
def ghosty(inventory, health):
    print("ðŸŒ‘ A Ghosty emerges! It drains your energy slowly.")
    damage = random.randint(1, 3)

    if "Cross" in inventory:
        print(f"You fight back with your cross! The Ghosty vanished")
    else:
        return health - damage
```

```python
def poltergeist(inventory, health):
    print("ðŸª‘ A Poltergeist threw a chair on you.")
    damage = random.randint(1, 4)

    if "Shield" in inventory:
        print(f"You protected yourself with the shield! The Poltergeist vanished")
    else:
        return health - damage
```

```python
def banshee(inventory, health):
    print("ðŸ“¢ A Banshee appeared")
    damage = random.randint(2, 4)

    if "Holy Water" in inventory:
        print(f"You threw the Holy Water on Banshee! The Banshee vanished")
    else:
        return health - damage
```

```python
def spook(inventory, health):
    print("ðŸ‘» A Spooky appeared")
    damage = round(health * random.uniform(0, 0.5), 2)

    if "Salt" in inventory:
        print(f"You threw the Salt on Spooky! The Spooky vanished")
    else:
        return health - damage
```

```python
def encounter_ghost(inventory, health):
    print("You encounter a ghost!")
    # Create the ghosts list
    ghost_types = ["Banshee", "Ghosty", "Poltergeist", "Spooky"]
    # pick a random ghost
    pick_ghost = random.choice(ghost_types)

    # Create the list of items in the bad
    item = ["Cross", "Shield", "Holy Water", "Salt"]
    # pick a random item
    pick_item = random.choice(item)
    inventory.append(pick_item)

    if pick_ghost == "Banshee":
        health = banshee(inventory, health)
        print(f"you found the item {pick_item}")
    elif pick_ghost == "Poltergeist":
        health = poltergeist(inventory, health)
        print(f"you found the item {pick_item}")
    elif pick_ghost == "Ghosty":
        health = ghosty(inventory, health)
        print(f"you found the item {pick_item}")
    elif pick_ghost == "Spooky":
        health = spook(inventory, health)
        print(f"you found the item {pick_item}")

    return health
```
